```python
# risk_scoring_analysis.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

# Add src folder to path for importing project modules
sys.path.append('../src')

from data_fetcher import load_wallet_ids, load_transactions_for_wallet
from feature_engineering import extract_features_from_transactions
from normalization import normalize_and_score

# Load wallet list
wallet_csv_path = "../data/raw/Wallet_id.csv"
wallets = load_wallet_ids(wallet_csv_path)
print(f"Loaded {len(wallets)} wallets.")

# Load transactions JSON path
tx_json_path = "../data/raw/user-wallet-transactions.json"

# Extract features for each wallet (including additional features)
feature_list = []
for wallet in wallets:
    txs = load_transactions_for_wallet(wallet, tx_json_path)
    features = extract_features_from_transactions(wallet, txs)
    if features:
        feature_list.append(features)

# Create DataFrame of extracted features and set wallet as index
features_df = pd.DataFrame(feature_list).set_index('wallet')

# Display sample extracted features with more columns
print("Sample features data:")
display(features_df.head())

# Apply normalization and calculate risk scores
scores_df = normalize_and_score(features_df)

# Display risk score summary statistics
print("\nRisk score summary stats:")
display(scores_df.describe())

# Merge scores with features for combined view
merged_df = features_df.join(scores_df)

print("\nTop 10 wallets by risk score:")
display(merged_df.sort_values('risk_score', ascending=False).head(10))

# Define risk categories for exploratory plots
def risk_category(score):
    if score <= 300:
        return 'Low Risk'
    elif score <= 700:
        return 'Medium Risk'
    else:
        return 'High Risk'

merged_df['risk_category'] = merged_df['risk_score'].apply(risk_category)

# Plot 1: Risk Score Distribution
plt.figure(figsize=(10, 6))
sns.histplot(merged_df['risk_score'], bins=30, kde=True, color='red')
plt.title('Wallet Risk Score Distribution')
plt.xlabel('Risk Score (0 = Low risk, 1000 = High risk)')
plt.ylabel('Number of Wallets')
plt.savefig("../outputs/risk_distribution.png")
plt.show()

# Plot 2: Correlation Heatmap of Features
numeric_df = merged_df.select_dtypes(include=['number'])
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Feature Correlation Heatmap')
plt.savefig("../outputs/feature_correlation_heatmap.png")
plt.show()

# Plot 3: Scatter plots of Risk Score vs Liquidation Count & Borrow Frequency
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(data=merged_df, x='liquidation_count', y='risk_score', hue='risk_category', palette='Set1')
plt.title('Risk Score vs Liquidation Count')

plt.subplot(1, 2, 2)
sns.scatterplot(data=merged_df, x='borrow_frequency', y='risk_score', hue='risk_category', palette='Set1')
plt.title('Risk Score vs Borrow Frequency')

plt.tight_layout()
plt.savefig("../outputs/scatter_risk_liquidation_borrow.png")
plt.show()

# Plot 4: Boxplot of Liquidation Count by Risk Category
plt.figure(figsize=(10, 6))
sns.boxplot(x='risk_category', y='liquidation_count', data=merged_df, palette='Set2')
plt.title('Liquidation Count by Risk Category')
plt.savefig("../outputs/boxplot_liquidation_by_risk_category.png")
plt.show()

# Plot 5: Pairplot of key features colored by risk category
pairplot_fig = sns.pairplot(merged_df,
             hue='risk_category',
             vars=['liquidation_count', 'borrow_frequency', 'repayment_ratio', 'risk_score'],
             palette='Set1',
             plot_kws={'alpha':0.7})
pairplot_fig.fig.suptitle('Pairplot of Features by Risk Category', y=1.02)
pairplot_fig.savefig("../outputs/pairplot_features_by_risk_category.png")
plt.show()

# Plot 6: Top 10 high risk wallets bar chart
top_n = 10
top_wallets = merged_df.nlargest(top_n, 'risk_score')

plt.figure(figsize=(12, 6))
sns.barplot(x=top_wallets.index, y=top_wallets['risk_score'], palette='Reds_r')
plt.xticks(rotation=45, ha='right')
plt.title(f'Top {top_n} High Risk Wallets')
plt.ylabel('Risk Score')
plt.tight_layout()
plt.savefig("../outputs/top_10_high_risk_wallets_bar.png")
plt.show()
```
